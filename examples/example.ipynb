```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from feature_importance_sdk import FeatureImportanceSDK
import sklearn
```

```python
# Load dataset
from sklearn.datasets import load_iris
data = load_iris()
X, y = data.data, data.target
feature_names = data.feature_names

# Initialize the model and SDK
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
sdk_instance = FeatureImportanceSDK(model)

# Calculate feature importance
importances = sdk_instance.calculate_importance(X, y)
print("Feature Importances:", importances)

# Plot feature importance
sdk_instance.plot_importance(feature_names, 'orange')
```
